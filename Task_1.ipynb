```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_56.csv'
df = pd.read_csv(file_path, skiprows=4)

```

### **1.POPULATION GROWTH IN INDIA**

```python
# Clean up the dataset by selecting relevant columns
df_clean = df[['Country Name', '1960', '1970', '1980', '1990', '2000', '2010', '2020']]

country_data = df_clean[df_clean['Country Name'] == 'India'].iloc[0, 1:]

# Plot the population trend
plt.figure(figsize=(10, 6))
country_data.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Population Growth in India (1960-2020)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **2.POPULATION GROWTH FOR MULTIPLE COUNTRIES OVER TIME**

```python
# Filter data for specific countries
countries = ['India', 'China', 'United States']
multi_country_data = df_clean[df_clean['Country Name'].isin(countries)]

# Set 'Country Name' as the index for better plotting
multi_country_data.set_index('Country Name', inplace=True)

# Transpose the data for a grouped bar chart
multi_country_data_transposed = multi_country_data.T

# Plot the grouped bar chart
multi_country_data_transposed.plot(kind='bar', figsize=(12, 6))
plt.title('Population Growth Comparison (1960-2020)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.legend(title='Country', fontsize=12)
plt.tight_layout()
plt.show()
```

### **3.HISTOGRAM OF POPULATION FOR A SINGLE YEAR**

```python
# Select population data for 2020
population_2020 = df_clean['2020']

# Drop missing values
population_2020 = population_2020.dropna()

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_2020, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Across Countries (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()
```

### **4.LINE CHART FOR POPULATION TRENDS**

```python
# Extract population data for India
india_data = df_clean[df_clean['Country Name'] == 'India'].iloc[0, 4:-1]

# Plot the line chart
plt.figure(figsize=(12, 6))
india_data.plot(kind='line', marker='o', color='green', linewidth=2)
plt.title('Population Trend in India (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **5.TOP 10 MOST POPULOUS COUNTRIES**

```python
# Sort countries by population in 2020
top_countries_2020 = df_clean[['Country Name', '2020']].sort_values(by='2020', ascending=False).head(10)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_countries_2020['Country Name'], top_countries_2020['2020'], color='orange')
plt.title('Top 10 Most Populous Countries (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.tight_layout()
plt.show()
```

### **6.PIE CHART OF GLOBAL POPULATION DISTRIBUTION**

```python
# Select data for specific countries
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']
global_data = df_clean[df_clean['Country Name'].isin(selected_countries)]
global_data_2020 = global_data[['Country Name', '2020']].set_index('Country Name')

# Add a "Rest of the World" category
rest_of_world = df_clean['2020'].sum() - global_data_2020['2020'].sum()
global_data_2020.loc['Rest of the World'] = rest_of_world

# Plot the pie chart
plt.figure(figsize=(8, 8))
global_data_2020.plot(kind='pie', y='2020', autopct='%1.1f%%', legend=False, startangle=140)
plt.title('Global Population Distribution (2020)', fontsize=16)
plt.ylabel('')  # Remove y-axis label
plt.tight_layout()
plt.show()
```

```python
# Drop rows with missing values in the "2020" column
df_clean_2020 = df_clean[['Country Name', '2020']].dropna()

# Select data for specific countries
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']
global_data = df_clean_2020[df_clean_2020['Country Name'].isin(selected_countries)]

# Set the index and calculate the "Rest of the World" category
global_data_2020 = global_data.set_index('Country Name')
rest_of_world = df_clean_2020['2020'].sum() - global_data_2020['2020'].sum()
global_data_2020.loc['Rest of the World'] = rest_of_world

# Verify the data
print(global_data_2020)

# Plot the pie chart with enhanced readability
plt.figure(figsize=(8, 8))
colors = ['gold', 'lightblue', 'coral', 'limegreen', 'violet', 'purple']  # Add distinct colors
explode = [0.1] + [0] * 4 + [0.1]  # Highlight the largest segment (China)

global_data_2020['2020'].plot(
    kind='pie',
    autopct=lambda p: f'{p:.1f}%\n({p * global_data_2020["2020"].sum() / 100:,.0f})',  # Show percentage and value
    startangle=140,
    colors=colors,
    explode=explode,  # Explode specific slices
    wedgeprops=dict(edgecolor='black')  # Add border for better visibility
)


plt.title('Global Population Distribution (2020)', fontsize=16)
plt.ylabel('')  # Remove y-axis label
plt.legend(
    global_data_2020.index,
    title='Regions',
    loc='center left',
    bbox_to_anchor=(1, 0.5),  # Place legend outside the chart
    fontsize=10
)
plt.tight_layout()
plt.show()

```
