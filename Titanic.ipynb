```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = ("/content/gender_submission.csv")
data = pd.read_csv(file_path)

# Data Cleaning
# 1. Check for missing values
missing_values = data.isnull().sum()

# 2. Check for duplicate rows
duplicates = data.duplicated().sum()

# 3. Verify unique value counts for each column
unique_counts = data.nunique()

# Display data summary
print("Missing Values:\n", missing_values)
print("\nDuplicates:", duplicates)
print("\nUnique Value Counts:\n", unique_counts)

# Exploratory Data Analysis
# 1. Visualize the distribution of the 'Survived' column
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=data, palette='viridis')
plt.title('Survival Distribution', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 2. Analyze survival rate
survival_rate = data['Survived'].value_counts(normalize=True) * 100
print("\nSurvival Rate:\n", survival_rate)

# 3. Plot survival rate as a pie chart
plt.figure(figsize=(8, 6))
data['Survived'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'orange'],
    labels=['Not Survived', 'Survived']
)
plt.title('Survival Rate', fontsize=16)
plt.ylabel('')  # Hide y-label for better visualization
plt.show()

# Identify Patterns and Trends
# Since the dataset only contains two variables (PassengerId and Survived),
# we can look for patterns in survival using grouping or aggregations if other attributes are present in an extended dataset.
print("\nDataset is limited to two variables. Add more features to explore patterns further.")
```
